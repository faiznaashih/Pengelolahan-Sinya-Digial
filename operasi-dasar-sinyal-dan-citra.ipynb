# Pengolahan Sinyal dan Citra - Operasi Dasar

## Tujuan
Menerapkan operasi dasar pada sinyal dan citra digital, seperti penjumlahan, penskalaan, pergeseran, dan refleksi. Proyek ini menggunakan sinyal audio dan citra produk sebagai studi kasus.

## Deskripsi Dataset
- **Sinyal**: Contoh suara .wav dari librosa.
- **Citra**: Gambar produk dari internet.


```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import cv2
from PIL import Image
import urllib.request
from IPython.display import Audio
```

```python
audio_path = librosa.example('trumpet')
x, sr = librosa.load(audio_path)

plt.figure(figsize=(12, 3))
librosa.display.waveshow(x, sr=sr)
plt.title("Sinyal Asli")
plt.show()

Audio(x, rate=sr)
```

```python
image = np.array(Image.open("/kaggle/input/trumpet/trumpet.jpeg").convert("RGB"))

plt.imshow(image)
plt.title("Gambar trumpet")
plt.axis('off')
plt.show()
```

```python

x_sum = x + x

x_scaled = 0.5 * x

delay = 1000
x_shifted = np.pad(x, (delay, 0))[:len(x)]

x_flipped = x[::-1]
```

```python

image_sum = cv2.addWeighted(image, 0.5, image, 0.5, 0)

image_scaled = cv2.convertScaleAbs(image, alpha=1.5, beta=0)

rows, cols, _ = image.shape
M = np.float32([[1, 0, 50], [0, 1, 30]]) 
image_shifted = cv2.warpAffine(image, M, (cols, rows))

image_flipped = cv2.flip(image, 1)
```

```python
def show_wave(title, y):
    plt.figure(figsize=(12, 2))
    librosa.display.waveshow(y, sr=sr)
    plt.title(title)
    plt.show()

show_wave("Sinyal Asli", x)
show_wave("Penjumlahan Sinyal", x_sum)
show_wave("Penskalaan Amplitudo", x_scaled)
show_wave("Pergeseran Sinyal", x_shifted)
show_wave("Refleksi Sinyal", x_flipped)
```

```python
def show_image(title, img):
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

show_image("Asli", image)
show_image("Penjumlahan", image_sum)
show_image("Brightness x1.5", image_scaled)
show_image("Pergeseran", image_shifted)
show_image("Refleksi Horizontal", image_flipped)
```

# Analisis

## Sinyal
- Penjumlahan membuat amplitudo sinyal dua kali lipat.
- Penskalaan menurunkan volume suara.
- Pergeseran menyebabkan delay.
- Refleksi membalik sinyal (efek waktu mundur).

## Citra
- Penjumlahan tidak mengubah banyak karena nilainya sama.
- Penskalaan memperjelas gambar tapi bisa overexposed.
- Pergeseran menggeser objek.
- Refleksi menciptakan efek cermin.
